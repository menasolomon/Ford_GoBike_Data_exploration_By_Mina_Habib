```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv('Feb_GoBike_2019_Organised.csv')
```

## <h0><center><font color=' #0e6251'>Ford GoBike Usage by Feb Days Hours, Weekdays & Average Distance For</font></center></h0> 
## <h0><center><font color=' springgreen'>Subscribers</font> <font color=' #0e6251'>&</font> <font color=' steelblue'>Customers</font></center></h0>
    

<font color='#1c2833'>February Days Hours Gobike usage is showing us that most people who uses GoBike System as Subscriber are using it for their daily commutes "7am to 9am","4pm to 6pm" & less distance, but on the other hand the Customer slightly have increse in same hours which hasn't much effect & when we see Weekends Days that have less Bike rides fo Subscribers but almost the same for Customers so we can assume that Customers bike rides are for more distance so it could be for having fun or just escape bad traffics.</font>

```python
# Pie chart for User Type illustration
sb.set(font_scale=1.6)
pie, ax = plt.subplots(figsize=[11.69, 8.27])
colors2 = ['springgreen', 'steelblue']
labels = df.user_type.value_counts().keys()
plt.pie(x=df.user_type.value_counts(), autopct="%.1f%%", explode=[0.05]*2, labels=labels, pctdistance=0.5, colors = colors2)
plt.title("Delivery Tips by type", fontsize=20, y=0.95, fontweight='bold');
```

```python
# Hourly bike rides for february by user type
hours_per_user_count = df.groupby(["start_time_hour", "user_type"]).size().reset_index()
plt.figure(figsize=(11.69, 8.27))
palette2 = {'Subscriber': 'steelblue', 'Customer': 'springgreen'}
axis = sb.pointplot(x='start_time_hour', y=0 , hue='user_type', palette=palette2, scale=1, data = hours_per_user_count)
plt.title('Feb day Hours Trips per user type', fontsize=20, y=1.02, fontweight='bold')
plt.xlabel('Hours for All feb days')
plt.ylabel('Trips')
leg = axis.legend()
leg.set_title('User Type',prop={'size':15})
axis = plt.gca();
```

```python
# weekdays bike rides for february by user type
weekday_per_user_count = df.groupby(["start_time_weekday", "user_type"]).size().reset_index()
plt.figure(figsize=(11.69, 8.27))
weekdays = ['Mon','Tue','Wed','Thu','Fri', 'Sat', 'Sun']
axis = sb.pointplot(x='start_time_weekday', y=0 , hue='user_type', palette=palette2, scale=1, data = weekday_per_user_count, order=weekdays)
plt.title('The WeekDay Trips per user type', fontsize=20, y=1.02, fontweight='bold')
plt.xlabel('WeekDays')
plt.ylabel('Trips')
leg = axis.legend()
leg.set_title('User Type',prop={'size':13})
axis = plt.gca();
```

```python
# Average Trip distance by user type
plt.figure(figsize=(11.69, 5))
axis = df.groupby('user_type')['dist_km'].mean().plot(kind='barh', color=colors2)
axis.set_title('Average Trip distance by user type', fontsize=20, y=1.02, fontweight='bold')
axis.set_ylabel('User type')
axis.set_xlabel('Mean Trip distance by user type');
```

## <h><center><font color='#0e6251'> Ford GoBike Usage Comparison For</font></center></h> 
## <h><center><font color='mediumspringgreen'>17 - 26</font> <font color='#0e6251'>&</font> <font color='dodgerblue'>26 - 38</font> <font color='#0e6251'>&</font> <font color='blue'>38 - 50</font> <font color='#0e6251'>&</font> <font color='slateblue'>50 - 60</font></center></h>
    
<font color='#1c2833'>"26 - 38" age group are the main users of GOBike System especially  as subscribers thurthdays is the most Crowded Day, also "26-38" age group has the hgiest Distance & Speed Average. for "50-60" age group has many varians in average of distance and speed due to thier health, sudays for this age group is for home with grandchildern of resting. and  Weekends have for all age groups lowest distance, Speed & rides</font>

```python
# Pie Ploting Users due to age 
colors = ['dodgerblue', 'mediumspringgreen', 'blue' , 'slateblue']
pie, ax = plt.subplots(figsize=[11.69, 8.27])
labels = df.age_groups.value_counts().keys()
plt.pie(x=df.age_groups.value_counts(), autopct="%.1f%%", explode=[0.05]*4, labels=labels, pctdistance=0.5, colors=colors)
plt.title("GoBike Users Age Groups", fontsize=20, y=1.02, fontweight='bold');
```

```python
# Daily usege of the bike sharing system per user type and gender
plt.figure(figsize=(11.69, 8.27))
age_order = ['17 - 26', '26 - 38', '38 - 50', '50 - 60']
palette = {'17 - 26': 'mediumspringgreen', '26 - 38': 'dodgerblue', '38 - 50': 'blue', '50 - 60': 'slateblue'}
weekdays = ['Mon','Tue','Wed','Thu','Fri', 'Sat', 'Sun']
g = sb.catplot(data=df, x='start_time_weekday', col="user_type", hue="age_groups",hue_order=age_order, kind='count', sharey = False, palette=palette, order=weekdays)
g.set_axis_labels("Day", "Trips")
g._legend.set_title('Age Groups')
g.set_titles("{col_name}")
g.fig.suptitle('Daily usage of the bike share system per user type and Age', y=1.05, fontsize=20, fontweight='bold');
```

```python

weekday_per_user_count = df.groupby(['start_time_weekday', 'age_groups', 'dist_km']).size().reset_index()
plt.figure(figsize=(11.69, 8.27))
axis = sb.pointplot(x='start_time_weekday', y='dist_km' , hue='age_groups', palette=palette, scale=1, data = weekday_per_user_count, order=weekdays, legend_out=False)
plt.title('Average Distance Per Age Group In Weekdays', fontsize=20, y=1.02, fontweight='bold')
plt.xlabel('WeekDays')
plt.ylabel('Distance in KM')
leg = axis.legend()
leg.set_title('Age Groups',prop={'size':14})
axis = plt.gca();
```

```python

weekday_per_user_count = df.groupby(['start_time_weekday', 'age_groups', 'speed_kmh']).size().reset_index()
plt.figure(figsize=(11.69, 8.27))
axis = sb.pointplot(x='start_time_weekday', y='speed_kmh' , hue='age_groups', palette=palette, scale=1, data = weekday_per_user_count, order=weekdays, legend_out=False)
plt.title('Average Speed Per Age Group In Weekdays', fontsize=20, y=1.02, fontweight='bold')
plt.xlabel('WeekDays')
plt.ylabel('Speed in km/h')
leg = axis.legend()
leg.set_title('Age Groups',prop={'size':14})
axis = plt.gca();
```

## <h><center><font color='#0e6251'> Ford GoBike Usage Comparison For</font></center></h> 
## <h><center><font color='aqua'>Females</font> <font color='#0e6251'>&</font> <font color='steelblue'>Males</font> <font color='#0e6251'>&</font> <font color='fuchsia'>Others</font> </center></h>


<font color='#1c2833'> Males are 74.4% of Go Bike Users their average distance & Trip Duration are the least but speed is the Highest. Females are 23.5% of Go Bike Users their average speed is lowest for all age groups & others @ age "50-60" are the slowest  </font> 

```python
# Pie Ploting Users due to age groups
colors = ['steelblue', 'aqua', 'fuchsia']
pie, ax = plt.subplots(figsize=[11.69, 8.27])
labels = df.member_gender.value_counts().keys()
plt.pie(x=df.member_gender.value_counts(), autopct="%.1f%%", explode=[0.05]*3, labels=labels, pctdistance=0.5, colors=colors)
plt.title("GoBike Users Genders", fontsize=20, y=1.02, fontweight='bold');
```

```python
# Daily usege of the bike sharing system per user type and gender
sb.set(font_scale=1.6)
palette1 = {'Female': 'aqua', 'Male': 'steelblue', 'Other': 'fuchsia'}
g = sb.catplot(data=df, x='start_time_weekday', col="user_type", hue="member_gender", kind='count',palette= palette1, order=weekdays, sharey = False)
g.set_axis_labels("Day", "Trips")
g._legend.set_title('Gender')
g.set_titles("{col_name}")
g.fig.suptitle('Daily usage of the bike share system per user type and gender', y=1.02, fontsize=20, fontweight='bold');
```

```python
#heatmaping Distance by age groups and Genders
sb.set(font_scale=1.8)
fig, ax = plt.subplots(figsize=(11.69, 8.27))
age_user_speed = df.groupby(['age_groups', 'member_gender']).mean()['dist_km']
age_user_speed = age_user_speed.reset_index(name = 'distance_avg')
age_user_speed = age_user_speed.pivot(index = "member_gender", columns = 'age_groups', values = 'distance_avg')
sb.heatmap(age_user_speed, annot = True, fmt = '.2f',cmap="winter", ax=ax, linewidths=20, cbar_kws = {'label' : 'mean(distance_avg)'}, annot_kws={"fontsize":15})
ax.set_title('Average Distance of Age groups By Genders', y=1.02, fontsize=20, fontweight='bold');
```

```python
#heatmaping Trip Duration by age groups and Genders
fig, ax = plt.subplots(figsize=(11.69, 8.27))
age_user_speed = df.groupby(['age_groups', 'member_gender']).mean()['duration_min']
age_user_speed = age_user_speed.reset_index(name = 'duration_min_avg')
age_user_speed = age_user_speed.pivot(index = "member_gender", columns = 'age_groups', values = 'duration_min_avg')
sb.heatmap(age_user_speed, annot = True, fmt = '.2f',cmap="summer", ax=ax, linewidths=5, cbar_kws = {'label' : 'mean(duration_min_avg)'}, annot_kws={"fontsize":15})
ax.set_title('Average Trip Duration for Age groups By Genders', y=1.02, fontsize=20, fontweight='bold');
```

```python
#heatmaping Speed "Km/h" by age groups and Genders
fig, ax = plt.subplots(figsize=(11.69, 8.27))
age_user_speed = df.groupby(['age_groups', 'member_gender']).mean()['speed_kmh']
age_user_speed = age_user_speed.reset_index(name = 'speed_km/h_avg')
age_user_speed = age_user_speed.pivot(index = "member_gender", columns = 'age_groups', values = 'speed_km/h_avg')
sb.heatmap(age_user_speed, annot = True, fmt = '.2f',cmap="coolwarm", ax=ax, linewidths=5, cbar_kws = {'label' : 'mean(speed_km/h_avg)'}, annot_kws={"fontsize":15})
ax.set_title('Average Speed of Age groups By Genders', y=1.02, fontsize=20, fontweight='bold');
```
